    II) Calc√∫le as Raizes da seguinte fun√ß√£o do segundo grau:
   a) $f(x) = x^{2} + x -2$<br>
   b) $g(x) = x^{2} - 10x + 24$<br>

Ex√≠stem in√∫meras formas de calcular as raizes de uma fun√ß√£o do 2¬∞ Grau, as raizes s√£o justamente os pontos onde a fun√ß√£o corta o eixo **X**, geralmente quando se tem duas raizes, a fun√ß√£o corta em dois pontos, com uma raiz somente em um ponto no eixo **x** e sem raizes a par√°bola esta flutuando em algum quadrante tendo apenas nos n√∫meros complexos.

Uma das ferramentas para calcular as raizes √© justamente o Bh√°skara, dado pela f√≥rmula: $x = \frac{-b \pm \sqrt{\Delta}}{2 \cdot a}$ e o Delta a f√≥rmula √© $\Delta = b^{2} - 4 \cdot a \cdot b$ e para os v√©rtices √© dever de casa :D

Logo a resp√≥sta para a pergunta **a)**  utilizando bh√°skara √©:<br>
$\sqrt{\Delta = 1^{2} - 4 \cdot 1 \cdot -2}$ Logo o delta √©: $\sqrt{9}$: **3**  <br><br>
$x = \frac{-1 \pm 3}{1 \cdot 2}$ Logo as raizes s√£o: $-2$ e $1$

Por√©m n√£o √© so essa a forma de calcular a equa√ß√£o do 2¬∞ grau, existe a f√≥rmula dos produtos not√°veis, para isso, vamos seguir esses passos:<br>
$x^{2} + \frac{bx}{a} + \frac{c}{a} = 0$, Dividir todos os lados da equa√ß√£o por $a$.<br> Ap√≥s a divis√£o, transformar em um produto not√°vel e movimentar a $\frac{c}{a}$ para o outro lado, ent√£o:<br>
$x + (\frac{b}{2a})^{2} = -\frac{c}{a} + (\frac{b}{2a})^{2}$ Para transformar em produto not√°vel, foi uilizado a seguinte f√≥rmula: $(x + \frac{b}{2a})^{2} = x^2 + \frac{2b}{2a}x + (\frac{b}{2a})^{2}$ o primeiro a quadrado, mais duas vezes o primeiro vezes o segundo e o quadrado do segundo, por√©m n√£o tem o segundo termo ao quadrado, para isso foi preciso colocar ele ali, mas tem que colocar do outro lado tamb√©m, simplificando, temos ent√£o:<br>
$x + \frac{b}{2a} = \pm \frac{\sqrt{\Delta}}{2a}$ J√° foi tirado a raiz tamb√©m e simplificado para $\Delta$ a express√£o $b^{2} - 4 \cdot a \cdot c$, e como requer o m√≥dulo na raiz, por isso o $\pm$<br>
Finalizando nossa formula ficaria: $x = \frac{-b \pm \sqrt{\Delta}}{2a}$

E oque tudo isso quer nos dizer, simples, pode ser poss√≠vel transformar qualquer equa√ß√£o do 2¬∞ grau em um produto not√°vel quando poss√≠vel, geralmente vai facilitar as manipual√µes.

**b)** Para a alternativa B, vai ser utilizado o Produto Not√°vel.<br>
$x^{2} - 10x + 24 = 0$ pode ser escrita como $(x - 5)^{2} = \pm 1$, basta mover o $+24$ para o outro lado e somar $25$ em ambos os lados para aparecer o produto not√°vel, pois $(x - 5)^{2} = x^{2} + 2 \cdot 5 x + 5^{2}$<br>
Ent√£o, passando o $5$ para o outro lado, consegue as duas raizes, sendo $x_1 = 6$ e $x_2 = 4$

**c)** Para a alternativa C, vai ser utilizado a Fatora√ß√£o.<br>
$h(x) = x^{2} - 15x$ pode ser escrita como $h(x) = x(x - 15)$ basta fazer a distributiva para encontrar o mesmo resultado.<br>Lembra que eu tinha citado que o $c$ √© o ponto em que a par√°bola corta o eixo Y, logo essa par√°bola esta flutuando em um quadrante, sendo assim uma de suas raizes √© **0**, e a outra √© **+15**.

## 1.2. Limites

Limites nada mais √© que um conceito para se estudar em mais detalhes sobre o infinito, vou passar apenas a explica√ß√£o intuitiva sobre os limites.<br>
Como que uma fun√ß√£o se comporta dado um certo valor , √© √∫til quando se tem um polin√¥mio a saber qual √© o termo dominante na express√£o, a no√ß√£o mais b√°sica √© que quanto mais eu me aproximo de um certo valor, n√£o importa quanto eu me aproxime eu nuca vou chegar no destino, tipo um zoom infinito.

Imagina-se os pontos amerelhos estarem se aproximando do ponto em vermelho pela esquerda e os pontos em azuis pela direita o mais pr√≥ximo poss√≠vel dele mas nunca chegando nele, esse √© o conceito de limite, resumidamente, o objetivo √© chegar o mais pr√≥ximo poss√≠vel do ponto em vermelho mas nunca estar exatamente no ponto vermelho.

```julia
limit_example();
```

Ou seja, no Limite onde o meu ponto amarelo $a$ esta tendendo ou ficando cada vez mais pr√≥ximo do ponto em vermelho $v$ √© dado pela seguinte express√£o: $\lim_{a \to v} f(x)$ a palavra $lim$ da essa no√ß√£o, e o $a$ √© o n√∫mero ou ponto que esta sendo aproxim√°do para o outro ponto que no caso √© o vermelho $v$.<br> Matematicamente: Conforme o $a$ se aproxima do $v$ pera direita ou esquerda, a $f(a)$ se aproxima da $f(v)$.

A seguinte fun√ß√£o $f(x) = x + 1 $, no limite quando essa fun√ß√£o tende a 1, fica como no DataFrame, logo escrita, fica $\lim_{x \to 1} (x + 1)$, isso √© verdade para outros n√∫meros tamb√©m.

## 1.3. Derivadas

Resumidamente a derivada √© a reta tangente ao gr√°fico de uma fun√ß√£o em um ponto escolhido, nome grande para uma coisa simples.<br>Essa ferramenta esta ligada a otimiza√ß√£o, taxa de varia√ß√£o, onde uma vari√°vel depende de outra., ou seja, se eu tiver a velocidade e quiser a acelera√ß√£o, vai ser necess√°rio a deriva√ß√£o, resumidamente, v√°rios problemas caem na derivada.

E a reta tangente √© justamente uma reta em um ponto espec√≠fico que corta uma fun√ß√£o somente em uma parte, logo as demais retas que cortam em mais posi√ß√µes s√£o chamadas de retas secantes.

```julia
derivative();
```

Nessa imagem, estou aproximando um ponto qualquer ao ponto vermelho utilizando a no√ß√£o de limite.

### 1.3.1. Como surgiu a Derivada.

E com toda reta √© possivel desenhar um tri√¢ngulo e calcular a tangente dado os pontos, que √© a diferen√ßa da altura pela diferen√ßa da base, no gr√°fico acima, vamos calcular essa diferen√ßa.<br><br>
Ponto em Laranja: $tg = \frac{f(15) - f(5)}{15 - 5}$ a fun√ß√£o avaliada no ponto 15 - a fun√ß√£o avaliada no ponto 5 e a diferen√ßa, por√©m vai dar um erro grande essa diferen√ßa, pois n√£o √© exatamente um tri√¢ngulo, e sim uma fun√ß√£o exponencial, logo precisa de um tri√¢ngulo menor para conseguir representar essa fun√ß√£o que n√£o √© uma reta.<br>

Ponto em Verde: $tg = \frac{f(11) - f(4)}{11 - 5}$ mesmo assim ainda √© poss√≠vel pegar um triangulo menor, e no limite, temos:<br>

Ponto em Vermelho: $tg = \frac{f(5.00005) - f(5)}{5.00005 - 5}$, logo essa aqui esta mais perto da verdadeira mas √© necess√°rio formular no limite, pois alguem pode vim pegar um ponto menor, que consequentemente vai ser melhor e assim sucessivamente.

Ou seja, √© necess√°rio resumir, mas para resumir √© com o limite

$f'(v) = \lim_{p \to v} \frac{f(p) - f(v)}{p - v}$ No limite em que o os outros pontos se aproximan do ponto vermelho, existe essa fun√ß√£o, logo essa √© a derivada dessa fun√ß√£o no ponto vermelho $v$.<br>Mas, geralmente √© utilizada outra f√≥rmula:<br>$f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}$ essa f√≥rmula basicamente fala que depende do $h$ que seria o passo, ou seja, o passo indo a 0, partindo dai, surge a derivada com o limite.

Entenda esse passo como um n√∫mero proximo a o ponto $v$ por exemplo.

$f'(x) = \lim_{h \to 0} 50$ ou seja, a $f(x) = ax + b = f'(x) = a$.<br>

Por√™m existem algumas regras para facilitar a vida de grandes deriva√ß√µes.

### 1.3.2. Regras de Deriva√ß√£o.

$f(x) = \mathbb{N} $ **a regra √©** $f'(x) = 0$<br>
$f(x) = x^{n}$ **a regra √©** $f'(x) = n \cdot x^{n-1}$<br><br>
$f(x) = sin(x)$ **a regra √©** $f'(x) = cos(x)$<br>
$f(x) = cos(x)$ **a regra √©** $f'(x) = -sin(x)$<br><br>
$f(x) = e^{x}$ **a regra √©** $f'(x) = e^{x}$<br>
$f(x) = ln(x)$ **a regra √©** $f'(x) = \frac{1}{x}$<br><br>

1. Quando a fun√ß√£o √© um n√∫mero natural, a derivada dessa fun√ß√£o √© uma constante 0.
2. Quando a fun√ß√£o √© elevada a algum expo√™nte, basta descer o expo√™nte multiplicando o valor e subtrair (-1) do expo√™nte.
3. Quando a fun√ß√£o √© a seno, a derivada √© a coseno.
4. Quando a fun√ß√£o √© a coseno, a derivada √© a *- seno*.
5. Quando a fun√ß√£o √© uma exponencial, a derivada √© ela mesmo.
6. Quando a fun√ß√£o √© um log, a derivada √© $x^{-1}$.

Essas regras podem ser facilmente encontradas pela internet, e o mais import√¢nte, todas essas regras podem ser provadas com o Limite.

$(f(x) + g(x))'$ **a regra √©** $f'(x) + g'(x)$<br><br>
$(n \cdot f(x))'$ **a regra √©** $n \cdot f'(x)$<br><br>
$f(x) \cdot g(x)$ **a regra √©** $f'(x) \cdot g(x) + f(x) \cdot g(x)' $<br><br>
$f(x) \div g(x)$ **a regra √©** $\frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{g(x)^2} $<br><br>

### 1.3.3. Exemplos das Regras de Deriva√ß√£o.

$f(x) = 2x^{5} + ln(x) - sin(x)$ No primeiro elemento o $5$ cai multiplicando o $2$ e subtrai $1$ do expoente (Regra 2), o $ln(x)$ √© justamente $\frac{1}{x}$ e por fim, o $sin(x)$ √© o $cos(x)$.<br><br>
$f'(x) = 10x^{4} + \frac{1}{x} - cos(x)$

$(f(x) = x^{2}) \cdot (g(x) = sin(x))$ Nesse caso vai ser utilizado a regrinha. 
$f'(x) \cdot g(x) + f(x) \cdot g'(x)$ <br><br>
$(x^{2})' \cdot sin(x) + x^{2} \cdot (sin(x))'$ agora deriva individualmente, logo $ 2x \cdot sin(x) + x^{2} \cdot cos(x)$

$\frac{x^{2}}{sin(x)}$ ou $x^{2} \div sin(x)$<br><br>
$\frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{g(x)^{2}}$ mudando os elementos $\frac{(x^{2})' \cdot sin(x) - x^{2} \cdot (sin(x))'}{sin(x)^2}$<br><br>
$\frac{2x \cdot sin(x) - x^{2} \cdot cos(x)}{sin(x)^2}$ para simplificar pode separar o $sin(x)^{2}$ em duas divis√µes e cortar o $sin(x)$ de cima da fra√ß√£o.
